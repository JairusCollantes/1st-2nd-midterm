```python
import pandas as gyat
import hvplot.pandas

df = gyat.read_csv("covid.csv")

df[["Country/Region","Confirmed","Deaths", "Recovered"]]
```

## Question 1

### Which country has the highest confirmed cases

```python
df.sort_values(by = "Deaths", ascending = False)[["Country/Region", "Confirmed"]].head(1)
```

## Question 2

### Which country has the highest deaths

```python
df.sort_values(by = "Deaths", ascending = False)[["Country/Region", "Deaths"]].head(1)
```

## Question 3

### Show Philippines Confirmed , Deaths , Recovered

```python
df[df["Country/Region"] == "Philippines"][["Country/Region", "Confirmed", "Deaths", "Recovered"]]
```

## Question 4

### Display Per WHO Region

```python
df.groupby("WHO Region")[["Confirmed", "Deaths", "Recovered"]].sum()
```

## Question 5

### What is the average Deaths per WHO region

```python
df.groupby("WHO Region")[["Deaths"]].mean().round(2)
```

## HVPLOT 1

### Total cases per WHO region

```python
contint = df.groupby("WHO Region")[["Confirmed", "Deaths", "Recovered", "Active"]].sum()
contint["Confirmed"].hvplot.bar(title = "Total Confirmed Cases by WHO Region", ylabel = "Total Confirmed Cases", rot = 45)
```

## HVPLOT 2

### confirmed and deaths

```python
contint[["Confirmed","Deaths"]].hvplot.line(title = "Confirmed Cases VS Deaths per WHO region",xlabel = "WHO region", ylabel = "Number of Cases")
```
